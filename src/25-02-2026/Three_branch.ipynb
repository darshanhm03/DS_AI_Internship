1. Types of Machine Learning


1. Supervised Learning

Definition: Machine learning where the data comes with labels or answers. The model learns a mapping from input features to output labels.

Goal: Predict labels for new, unseen data.

Example tasks: Classification, regression.

2. Unsupervised Learning

Definition: Machine learning where the data has no labels. The model tries to find hidden patterns, structures, or clusters in the data.

Goal: Understand the underlying structure or distribution.

Example tasks: Clustering, dimensionality reduction, anomaly detection.
3.Reinforcement Learning

Definition: Machine learning where an agent learns by interacting with an environment, receiving rewards or penalties based on its actions.

Goal: Learn a sequence of actions that maximizes cumulative reward.

Example tasks: Game playing, robotics control, recommendation systems with feedback.
